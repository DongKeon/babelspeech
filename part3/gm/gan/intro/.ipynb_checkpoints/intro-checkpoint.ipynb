- 게임 이론(game theory)은 상호 의존적인 의사 결정에 관한 이론. 게임(game)이란 효용 극대화를 추구하는 행위자들이 일정한 전략을 가지고 최고의 보상을 얻기 위해 벌이는 행위를 말함

> - 원래 우리가 풀고자하는 GAN 문제는 다음과 같은 minimax problem입니다:
> - G∗ = minG maxD V(G,D)
> - 그런데 우리가 실제 학습을 할 때는 G와 D에 대한 update를 번갈아가며 해주기 때문에 Neural network의 입장에서는 이러한 minimax problem과 아래와 같은 maximin problem이 구별이 되지 않습니다:
> - G∗ = maxD minG V(G,D)
> 문제는 위와 같은 maximin problem의 경우에서 생깁니다. 수식의 안 쪽부터 살펴보겠습니다.

### 2-1 (Notebook) Generative Adversarial Network Intro
- 노트북에서는 MNIST 데이터 세트에 대해 훈련된 GAN를 구축할 것입니다.
- GAN의 기본 개념은 두 개의 네트워크 인 발전기 G와 판별자 D가 서로 경쟁한다는 것입니다. 생성기는 가짜 데이터를 판별 자로 전달합니다. discriminator는 또한 실제 데이터를보고 수신 된 데이터가 실제인지 또는 가짜인지 예측합니다.
- 생성기는 판별자를 속일 수 있도록 훈련되어 있으므로 가능한 한 실제 데이터와 유사한 데이터를 출력하려고합니다. 그리고 판별 자 (discriminator)는 어떤 데이터가 진짜이고 어떤 데이터가 가짜인지 파악하도록 훈련 받았습니다. 
- 무슨 일이 일어나는지는 발생기가 실제 데이터에서부터 판별 자까지 구별 할 수없는 데이터를 만드는 것을 배웁니다.

```python
%matplotlib inline

import pickle as pkl
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data')
```

### 2-2 Model Inputs
- 먼저 그래프의 입력 값을 만들어야합니다.  두 개의 입력이 필요합니다. 하나는 판별자 용이고 다른 하나는 생성기 용입니다. 여기서 우리는 discriminator input inputs_real과 generator input inputs_z를 호출 할 것입니다. 우리는 각각의 네트워크에 적절한 크기를 할당 할 것입니다.

```python
def model_inputs(real_dim, z_dim):
    inputs_real = tf.placeholder(tf.float32, (None, real_dim), name='input_real') 
    inputs_z = tf.placeholder(tf.float32, (None, z_dim), name='input_z')
    
    return inputs_real, inputs_z
```

- 여기에서는 네트워크를 구축 할 것입니다. 이 네트워크를 보편적인 함수 근사자로 만들려면 최소한 하나의 숨겨진 레이어가 필요합니다. 
- 그라디언트가 방해받지 않고 레이어를 통해 뒤로 흐를 수 있도록 A leaky ReLU를 사용해야합니다. A leaky ReLU는 음의 입력 값에 대해 작은 0이 아닌 출력이 있다는 점을 제외하면 일반 ReLU와 같습니다.

### 2-3 Variable Scope
- 여기서 우리는 두 가지 이유로 tf.variable_scope를 사용해야합니다. 첫째, 모든 변수 이름이 생성자로 시작하는지 확인하겠습니다. 
- 마찬가지로, 우리는 discriminator 변수 앞에 변수를 추가합니다. 이것은 나중에 우리가 별도의 네트워크를 훈련 할 때 도움이 될 것입니다.
- 이름을 설정하기 위해 tf.name_scope을 사용할 수도 있지만,이 네트워크를 다른 입력으로 재사용하고 싶습니다. 
- generator 경우, 우리는 그것을 훈련 할 것이지만, 우리가 훈련하고 훈련을 할 때 그것으로부터 샘플을 얻을 것입니다. 판별 기는 가짜 및 실제 입력 이미지간에 변수를 공유해야합니다. 따라서 tf.variable_scope에 reuse 키워드를 사용하여 TensorFlow에게 그래프를 다시 작성하는 경우 새 변수를 작성하는 대신 변수를 재사용하도록 지시 할 수 있습니다.
- tf.variable_scope를 사용하려면 with 문을 사용합니다.

### 2-4 leaky ReLU
- TensorFlow는 leaky ReLUs에 대한 작업을 제공하지 않으므로 이를 작성해야합니다. 
- 선형으로 완전히 연결된 레이어의 출력을 가져 와서 tf.maximum으로 전달할 수 있습니다. 
- 일반적으로 매개 변수 alpha는 음수 값의 출력 크기를 설정합니다. 따라서 음수 입력 (x) 값의 출력은 alpha * x이고 양수 x의 출력은 x : $$입니다.


### 2-5 Tanh 산출
- generator의 출력을 위해 tanh로 최고의 성능을 발휘합니다. 이것은 우리가 0과 1 대신에 -1과 1 사이가되도록 MNIST 이미지의 크기를 재조정해야 함을 의미합니다.

```python
def generator(z, out_dim, n_units=128, reuse=False, alpha=0.01):
    with tf.variable_scope('generator', reuse=reuse):
        # Hidden layer
        h1 = tf.layers.dense(z, n_units, activation=None)
        # Leaky ReLU
        h1 = tf.maximum(alpha * h1, h1)
        # Logits and tanh output
        logits = tf.layers.dense(h1, out_dim, activation=None)
        out = tf.tanh(logits)
        return out
```

### 2-6 Discriminator
- The discriminator network is almost exactly the same as the generator network, except that we're using a sigmoid output layer.

```python
def discriminator(x, n_units=128, reuse=False, alpha=0.01):
    with tf.variable_scope('discriminator', reuse=reuse):
        # Hidden layer
        h1 = tf.layers.dense(x, n_units, activation=None)
        # Leaky ReLU
        h1 = tf.maximum(alpha * h1, h1)
        logits = tf.layers.dense(h1, 1, activation=None)
        out = tf.sigmoid(logits)
        return out, logits
```

```python
# Size of input image to discriminator
input_size = 784
# Size of latent vector to generator
z_size = 100
# Sizes of hidden layers in generator and discriminator
g_hidden_size = 128
d_hidden_size = 128
# Leak factor for leaky ReLU
alpha = 0.01
# Smoothing 
smooth = 0.1
```

### 2-7 네트워크 구축
- 이제 위에서 정의한 함수로 네트워크를 구축하고 있습니다.
- 첫 번째는 input과 z의 크기를 사용하여 model_inputs에서 입력 인 input_real, input_z을 얻는 것입니다.
- 그런 다음 생성기 생성기 (input_z, input_size)를 생성합니다. 이렇게하면 적절한 입력 및 출력 크기로 생성기가 작성됩니다.
- 그렇다면 discriminators. 실제 데이터 용과 가짜 데이터 용으로 두 가지를 빌드 할 것입니다. 실제 데이터와 가짜 데이터에서 가중치를 동일하게 유지하려면 변수를 다시 사용해야합니다. 
- 가짜 데이터의 경우 생성기에서 g_model로 가져옵니다. 따라서 실제 데이터 판별 기는 판별자(input_real)이고 가짜 판별자는 판별 자 (g_model, reuse = True)입니다.

```python
tf.reset_default_graph()
# Create our input placeholders
input_real, input_z = model_inputs(input_size, z_size)

# Build the model
g_model = generator(input_z, input_size, n_units=g_hidden_size, alpha=alpha)
# g_model is the generator output

d_model_real, d_logits_real = discriminator(input_real, n_units=d_hidden_size, alpha=alpha)
d_model_fake, d_logits_fake = discriminator(g_model, reuse=True, n_units=d_hidden_size, alpha=alpha)
```

### 2-8 Discriminator and Generator Losses
- 이제 우리는 손실을 계산할 필요가 있습니다. 이는 약간 까다롭습니다. 
- discriminator의 경우 총 손실은 실제 이미지와 가짜 이미지의 손실 합계입니다 (d_loss = d_loss_real + d_loss_fake). Sigmoid cross-entropys에 의한 손실은 tf.nn.sigmoid_cross_entropy_with_logits로 얻을 수 있습니다. 또한 일괄 처리의 모든 이미지에 대한 평균을 얻기 위해 tf.reduce_mean에 랩핑 할 것입니다. 그래서 손실은 다음과 같이 보일 것입니다.
- tf.reduce_mean (tf.nn.sigmoid_cross_entropy_with_logits (logits = logits, labels = labels))

- 실제 이미지 로그에 대해서는 위의 셀에있는 discriminator에서 얻은 d_logits_real을 사용합니다. 
- 라벨은 모두 실제 이미지이므로 모든 라벨을 원하게 만듭니다. 판별자의 일반화를 돕기 위해 매개 변수 smooth를 사용하여 레이블을 1.0에서 0.9로 약간 줄입니다. 라벨 평활화는 일반적으로 분류 기준을 사용하여 성능을 향상시키는 데 사용됩니다. TensorFlow에서 label = tf.ones_like (텐서) * (1 - 부드러운)

- 위조된 데이터에 대한 판별 자 손실은 유사합니다. 로그는 d_logits_fake이며, 우리는 생성기 출력을 판별 자로 전달합니다. 
- 이 가짜 로그는 모두 0의 레이블과 함께 사용됩니다. 우리는 discriminator가 실제 이미지의 경우 1을 출력하고 위조 이미지의 경우 0을 출력하기를 원하므로 손실을 반영하도록 설정해야합니다.
- 마지막으로, Generator 손실은 가짜 이미지 로그 인 d_logits_fake를 사용하고 있습니다. 그러나 이제 레이블은 모두 1입니다. 생성기는 판별자를 속이려고 시도하기 때문에 가짜 이미지 용 판별자를 출력하려고합니다.


> GAN에서 일어나는 주요 failure mode중의 하나는 generator가 항상 하나의 포인트 값만을 생성하는 모델 패러미터 값에 빠져든다는 점, 이 문제를 해결하는 가장 확실한 방법은 discriminator가 여러 data example을 살펴보도록 해야 하는데, 이것을 minibatch discrimination이라고 한다.
> Label smoothing 방법은  classify하기위해 사용하는 label 0/1 targets을 덜 극단적인 값(0.9/0.1)으로 대체하는 것
> [improved-techniques-for-training-gans](https://jay.tech.blog/2016/12/15/improved-techniques-for-training-gans/)

```python
# Calculate losses
d_loss_real = tf.reduce_mean(
        tf.nn.sigmoid_cross_entropy_with_logits(logits=d_logits_real, 
        labels=tf.ones_like(d_logits_real) * (1 - smooth)))

d_loss_fake = tf.reduce_mean(
        tf.nn.sigmoid_cross_entropy_with_logits(logits=d_logits_fake, 
        labels=tf.zeros_like(d_logits_real)))

d_loss = d_loss_real + d_loss_fake

g_loss = tf.reduce_mean(
        tf.nn.sigmoid_cross_entropy_with_logits(logits=d_logits_fake,
        labels=tf.ones_like(d_logits_fake)))
```

- real_loss: 실제 이미지들을 넣은 결과값들(0 혹은 1로 구성된 벡터)와 실제 이미지들의 레이블(1로 이루어진 벡터)를 비교해서 계산된 loss (smooth를 줌)
- fake_loss: fake 이미지들을 넣은 결과값(0 혹은 1로 구성된 벡터)와 fake 이미지들의 레이블(영벡터)를 비교해서 계산된 loss
- Discriminator's loss = real_loss + fake_loss, 두 loss를 더해주면 Discriminator의 최종 loss가 됩니다. 이를 역전파하고 parameter를 update 합니다.

- fake 이미지들을 Discriminator에 넣음, Discriminator가 판단하기에 각각의 이미지가 실제 이미지일 확률이 반환
- Generator의 목적은 Discriminator가 가짜 이미지를 진짜 이미지라고 착각하게 만드는 것. 따라서 아까 계산된 확률값을 실제 이미지 label(1)로 구성된 벡터와 비교하여 Loss를 계산함. 
- 다시 말하자면, fake 이미지를 집어넣어놓고 마치 이게 실제 이미지였던 것 마냥 loss를 계산하고, 학습하는 것. 이러한 Generator의 loss를 최소화 하는 과정에서 fake 이미지들이 실제 이미지와 유사해집니다.

### 최적화 도구
- generator와 discriminator 변수를 따로 갱신하려고 합니다. 
    - 따라서 우리는 두 부분에 대한 각 부분 빌드 최적화 프로그램을 위한 변수를 가져와야합니다. 
    - 모든 훈련 가능한 변수를 얻으려면 tf.trainable_variables()를 사용합니다. 그래프에 정의 된 모든 변수 목록이 생성됩니다.

- 생성기 옵티 마이저의 경우 생성기 변수 만 필요합니다. 우리의 과거 자아는 좋았고 generator로 모든 generator 변수 이름을 시작하기 위해 가변 범위를 사용했습니다. 따라서 tf.trainable_variables ()의 목록을 반복하고 생성기로 시작하는 변수를 유지하면 됩니다. 
    - 각 변수 객체는 변수 이름을 문자열 (var.name == 'weights_0')로 유지하는 속성 이름을가집니다.

- 우리는 판별 자와 비슷한 것을 할 수 있습니다. discriminator의 모든 변수는 discriminator로 시작합니다.
    - 그런 다음 optimizer에서 변수 목록을 최소화 메소드의 var_list에 전달합니다. 이것은 옵티마이저가 나열된 변수만 갱신하도록 지시합니다.
    - tf.train.AdamOptimizer()와 같은 것입니다. minimize (loss, var_list = var_list)는 var_list의 변수 만 조정합니다.

```python
# Optimizers
learning_rate = 0.002

# Get the trainable_variables, split into G and D parts
t_vars = tf.trainable_variables()
g_vars = [var for var in t_vars if var.name.startswith('generator')]
d_vars = [var for var in t_vars if var.name.startswith('discriminator')]

d_train_opt = tf.train.AdamOptimizer(learning_rate).minimize(d_loss, var_list=d_vars)
g_train_opt = tf.train.AdamOptimizer(learning_rate).minimize(g_loss, var_list=g_vars)
```

```python
batch_size = 100
epochs = 100
samples = []
losses = []
# Only save generator variables
saver = tf.train.Saver(var_list=g_vars)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for e in range(epochs):
        for ii in range(mnist.train.num_examples//batch_size):
            batch = mnist.train.next_batch(batch_size)
            
            # Get images, reshape and rescale to pass to D
            batch_images = batch[0].reshape((batch_size, 784))
            batch_images = batch_images*2 - 1 #0 ~ 1 rescaling -> -1 ~ 1
            
            # Sample random noise for G
            batch_z = np.random.uniform(-1, 1, size=(batch_size, z_size))
            
            # Run optimizers
            _ = sess.run(d_train_opt, feed_dict={input_real: batch_images, input_z: batch_z})
            _ = sess.run(g_train_opt, feed_dict={input_z: batch_z})
        
        # At the end of each epoch, get the losses and print them out
        train_loss_d = sess.run(d_loss, {input_z: batch_z, input_real: batch_images})
        train_loss_g = g_loss.eval({input_z: batch_z})
            
        print("Epoch {}/{}...".format(e+1, epochs),
              "Discriminator Loss: {:.4f}...".format(train_loss_d),
              "Generator Loss: {:.4f}".format(train_loss_g))    
        # Save losses to view after training
        losses.append((train_loss_d, train_loss_g))
        
        # Sample from generator as we're training for viewing afterwards
        sample_z = np.random.uniform(-1, 1, size=(16, z_size))
        gen_samples = sess.run(
                       generator(input_z, input_size, n_units=g_hidden_size, reuse=True, alpha=alpha),
                       feed_dict={input_z: sample_z})
        samples.append(gen_samples)
        saver.save(sess, './checkpoints/generator.ckpt')
```

### 2-9 Training loss
- Here we'll check out the training losses for the generator and discriminator.

```python
fig, ax = plt.subplots()
losses = np.array(losses)
plt.plot(losses.T[0], label='Discriminator')
plt.plot(losses.T[1], label='Generator')
plt.title("Training Losses")
plt.legend()
```
