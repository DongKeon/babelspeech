### 1.1   Machine learning fundamentals
- 수영하는 법을 누군가에게 설명해 본 적이 있습니까? 리듬 관절의 움직임과 유체 패턴을 설명하는 것은 그 복잡성 면에서 압도적입니다.
- 마찬가지로 일부 소프트웨어 문제는 우리의 마음을 쉽게 감쌀 수 없을만큼 복잡합니다. 이를 위해 기계 학습은 사용할 도구 일뿐입니다.

- 단순한 관점에서, 전통적인 프로그래밍은 각 입력에 대해 결정론적 결과를 가정합니다. 
- 반면 기계학습은 입출력 대응이 잘 이해되지 않는 문제를 해결할 수 있습니다.

- 귀납적 학습이 항상 필요한 것은 아니기 때문에 전에 프로그래밍 할 때 그러한 접근법을 사용하지 않았을 수도 있습니다. 임의의 두 숫자의 합이 짝수인지 홀수인지를 결정하는 작업을 고려하십시오. 물론 수백만 개의 교육 사례 (그림 1.1에 요약되어 있음)를 사용하여 기계 학습 알고리즘을 교육하는 것을 상상할 수 있지만, 지나침을 확실히 알고 있습니다. 좀더 직접적인 접근 방식을 사용하면 쉽게 트릭을 수행 할 수 있습니다.

![img_01_01](./img/01_01.png)

- 예를 들어, 두 개의 홀수의 합은 항상 짝수입니다. 두 개의 홀수를 취해서 합산 한 다음 합이 짝수인지 확인하십시오. 이 사실을 직접 증명할 수있는 방법은 다음과 같습니다. 

- 모든 정수 n에 대해 수식 2n + 1은 홀수를 생성합니다. 더욱이, 임의의 홀수는 어떤 값 n에 대해 2n + 1로서 기록 될 수있다. 그래서 숫자 3은 2 (1) + 1로 쓰여질 수 있고 숫자 5는 2 (2) + 1로 쓸 수 있습니다.

- 그래서 우리가 2n + 1과 2m + 1의 두 개의 다른 홀수를 가지고 있다고 가정합시다. 여기서 n과 m은 정수입니다. 두 개의 홀수를 더하면 (2n + 1) + (2m + 1) = 2n + 2m + 2 = 2 (n + m + 1)이됩니다. 2 배가 짝수이기 때문에 이것은 짝수입니다.

- 마찬가지로 2 개의 짝수의 합이 짝수 인 것도 알 수 있습니다. 2m + 2n = 2 (m + n). 그리고 마지막으로 홀수의 짝수가 홀수 인 이유를 추론합니다. 2m + (2n + 1) = 2 (m + n) + 1 

![img_01_02](./img/01_02.png)

- 기계 학습을 전혀 사용하지 않으면 누군가가 당신에게 던지는 모든 쌍의 쌍으로 이 작업을 해결할 수 있습니다. 수학적 규칙을 직접 적용하기 만하면이 문제를 해결할 수 있습니다. 그러나 ML 알고리즘에서 우리는 내부 논리를 블랙 박스로 취급 할 수 있습니다. 내부에서 일어나는 논리가 해석하기가 분명하지 않을 수 있습니다.

### Parameters
- 결정되지 않은 값을 매개 변수라고 하며 설명을 모델이라고 합니다. 당신의 임무는 최상의 모델을 달성하기 위해 매개 변수를 조정하는 최선의 방법을 파악하기 위해 기존의 예를 관찰하는 알고리즘을 작성하는 것입니다. 


### Learning and Inference
- 오븐에서 일부 디저트를 구우려고 한다고 가정 해보십시오. 주방을 처음 사용하는 분이라면 올바른 조합과 재료의 완벽한 비율로 맛이 좋은 것을 만드는 데 며칠이 걸릴 수 있습니다. 조리법을 녹음함으로써 궁극의 시식 식사를 발견 한 경우 디저트를 빨리 반복하는 방법을 기억할 수 있습니다.

- 마찬가지로, 기계 학습은 조리법에 대한 이 아이디어를 공유합니다. 일반적으로 학습과 추론의 두 단계로 알고리즘을 검사합니다. 학습 단계의 목적은 특성 벡터라고 불리는 데이터를 기술하고 그것을 모델로 요약하는 것입니다. 모델은 우리의 제조법입니다.

- 특징 벡터는 데이터의 실질적인 단순화입니다. 실제 개체를 속성 목록으로 충분히 요약 한 것으로 생각할 수 있습니다. 학습 및 추론 단계는 데이터 대신에 특징 벡터에 직접 의존한다.

- 레시피를 다른 사람들이 공유하고 사용할 수 있는 것과 마찬가지로, 학습된 모델은 다른 소프트웨어에서도 재사용됩니다. 학습 단계는 가장 많은 시간을 필요로합니다. 알고리즘을 실행하는 데는 며칠 또는 몇 주가 걸리지 않을지라도 유용한 모델로 수렴하는 데 몇 시간이 걸릴 수 있습니다. 그림 1.4는 학습 파이프 라인을 개략적으로 보여줍니다.

### 1.2   Data representation and features

- 컴퓨터는 정교한 계산기 일 뿐이므로 기계학습 시스템에 공급하는 데이터는 (1) 벡터, (2) 행렬 또는 (3) 그래프와 같은 수학적 개체여야 합니다.

- 신차를 시장에 내놓을 때, 다양한 제조사와 모델 간의 사소한 세부 사항을 파악하는 것이 중요합니다. 결국, 당신이 수천 달러를 쓰려고 한다면, 부지런히 그렇게 할 수 있습니다. 당신은 가능성이 각 자동차에 대한 기능의 목록을 기록하고 앞뒤로 비교할 것입니다. 정렬 된 기능 목록을 기능 벡터라고합니다.

- 자동차를 구입할 때, 주행 거리를 비교하는 것이 무게와 같은 관심사와 관련성이없는 것을 비교하는 것보다 유리할 수 있습니다. 추적 할 기능의 수가 너무 적지 않고 너무 많지 않아야합니다. 측정 횟수와 비교할 측정 항목을 모두 선택하는 엄청난 노력을 피쳐 엔지니어링이라고합니다. 검사하는 기능에 따라 시스템 성능이 크게 달라질 수 있습니다. 추적 할 올바른 기능을 선택하면 약한 학습 알고리즘을 보완 할 수 있습니다.

- ML의 일반적인 경험 규칙은 더 많은 데이터가 더 나은 결과를 산출한다는 것입니다. 그러나 더 많은 기능을 보유하고 있어도 항상 동일한 것은 아닙니다. 반 직관적 인 경우, 추적하는 기능의 수가 너무 많으면 성능이 저하 될 수 있습니다. 학자들은 이 현상을 차원의 저주라고 부릅니다. 

- 모든 데이터의 공간을 대표 샘플로 채우려면 피쳐 벡터의 크기가 커짐에 따라 지수 적으로 더 많은 데이터가 필요합니다. 그 결과, 피쳐 엔지니어링은 ML에서 가장 중요한 문제 중 하나입니다.

### 1.3   Distance Metrics

- 구매하려는 잠재적인 자동차의 특징 벡터가 있는 경우에는 특징 벡터에 거리 함수를 정의하여 두 모델이 가장 유사하다고 판단 할 수 있습니다. 
- 객체 간의 유사성을 비교하는 것은 기계학습의 필수 요소입니다. 피쳐 벡터는 우리가 다양한 방식으로 피사체를 비교할 수 있도록 피사체를 표현할 수 있게 합니다. 표준 접근법은 공간상의 점을 생각할 때 가장 직관적 인 기하학적 해석인 유클리드 거리를 사용하는 것입니다.
- x = (x1, x2, ..., xn)과 y = (y1, y2, ..., yn)의 두 가지 특징 벡터가 있다고 가정 해 봅시다. 유클리드 거리 || x-y || 에 의해 계산됩니다.

![img_01_09](./img/01_09b_F1.png)

![img_01_09](./img/01_09b_F2.png)

- 학자들은 이것을 L2 표준이라고 부릅니다. 하지만 실제로는 많은 거리 함수 중 하나 일뿐입니다. 또한 L0, L1 및 L- 무한대 표준이 존재합니다. 이 모든 규범은 거리를 측정하는 유효한 방법입니다. 여기에 자세히 나와 있습니다.

- L0 노옴은 벡터의 0이 아닌 요소의 총 수를 계산합니다. 예를 들어 원점 (0, 0)과 벡터 (0, 5) 사이의 거리는 1입니다. 0이 아닌 요소가 하나뿐이기 때문입니다.

- L1 표준은 Σ |xn|으로 정의됩니다. L1 규준 하의 두 벡터 사이의 거리를 맨하탄 거리라고도합니다. 거리가 그리드를 형성하는 뉴욕 맨해튼과 같은 도심 지역에 살고 있다고 상상해보십시오. 한 교차로에서 다른 교차로까지의 최단 거리는 블록을 따라 있습니다. 유사성, 두 벡터 간의 L1 거리는 직교 방향을 따른다. 따라서 L1 표준에서 (0, 1)과 (1, 0) 사이의 거리는 2입니다.

### 1.4   Types of Learning

- 이제는 특징 벡터를 비교할 수 있으므로 실제 알고리즘에 데이터를 사용하는 데 필요한 도구가 있습니다. 기계 학습은 종종 감독 학습, 감독되지 않은 학습 및 강화 학습의 세 가지 관점으로 나뉩니다. 하나씩 하나씩 살펴 보겠습니다.

### 1.4.1 Supervised Learning
- 감독된 기계 학습 시스템은 우리가 모델이라고 부르는 유용한 이해를 개발하기 위해 분류 된 데이터를 필요로 합니다. 간단히 말해, 모델은 일부 데이터에 레이블을 할당하는 함수입니다. 이는 학습 데이터 세트라고하는 이전 예제를 참조로 사용합니다.

- 모델에 대해 이야기하는 편리한 방법은 수학 표기법을 사용하는 것입니다. x를 특징 벡터와 같은 데이터의 인스턴스라고 하자. x와 연관된 해당 레이블은 f(x)이입니다. 보통 변수 y = f(x)를 사용하기 때문에 더 빨리 쓸 수 있습니다. 사진을 통해 사람의 민족성을 분류하는 예에서 x는 다양한 관련 기능의 백 - 차원 벡터가 될 수 있으며 y는 다양한 민족성을 나타내는 몇 가지 값 중 하나입니다. y는 값이 거의 없으므로 이 모델을 분류 자라고합니다. y가 많은 값을 가져올 수 있고 값이 자연 순서가있는 경우 모델을 회귀 변수라고합니다.

- 모델의 예측을 g(x)로 나타냅니다. 때로는 모델을 조정하여 성능을 크게 바꿀 수도 있습니다. 모델에는 인간 또는 자동으로 튜닝 할 수있는 매개 변수가 있습니다. 벡터 θ를 사용하여 매개 변수를 나타냅니다. 모두 합치면, g (x|θ)는 모델을 더 완전하게 나타내고, "주어진 x의 θ"를 읽는다.

- 모델의 예측 g (x|θ)의 성공 여부는 지표 진리 y와 얼마나 잘 일치 하는가에 달려있다. 이 두 벡터 사이의 거리를 측정 할 방법이 필요합니다. 예를 들어, L2-norm은 두 벡터가 얼마나 가까운 거리에 있는지 측정하는 데 사용될 수 있습니다. 지상 진실과 예측 사이의 거리를 비용이라고합니다.

- 감독 된 기계 학습 알고리즘의 본질은 최소 비용을 초래하는 모델의 매개 변수를 파악하는 것입니다. 수학적으로 말하자면 모든 데이터 포인트 x ∈ X 중에서 비용을 최소화하는 θ *를 찾으려고 합니다. 이 최적화 문제를 형식화하는 한 가지 방법은 다음과 같습니다.

- 분명히 매개 변수 공간이라고도 하는 모든 가능한 θs의 조합을 강요하는 것은 결국에는 최적의 솔루션을 찾지만 용인 할 수 없는 런타임에 발생합니다. 기계 학습의 주요 연구는 이 매개 변수 공간을 효율적으로 검색하는 알고리즘 작성에 관한 것입니다. 첫 번째 알고리즘 중 일부는 그래디언트 디센트 (gradient descent), 시뮬레이트 어닐링 (simulated annealing) 및 유전 알고리즘을 포함합니다. 

- 일단 매개 변수가 다른 방법으로 학습되면 모델이 데이터에서 패턴을 얼마나 잘 포착했는지 계산할 수 있습니다. 경험 법칙은 훈련에 사용한 것과 동일한 데이터로 모델을 평가하는 것이 아닙니다. 대부분의 데이터는 훈련을 위해 사용하고 나머지는 테스트 용으로 사용하십시오. 예를 들어 100 개의 레이블이 지정된 데이터가 있는 경우 70 개를 무작위로 선택하여 모델을 훈련하고 나머지 30 개를 테스트 데이터 지정하여 테스트합니다.


### 1.4.2   Unsupervised Learning
- 자율 학습은 해당 레이블이나 응답없이 오는 모델링 데이터에 관한 것입니다. 원시 데이터에서 결론을 내릴 수 있다는 사실은 마술처럼 느껴집니다. 충분한 데이터가 있으면 패턴과 구조를 찾을 수 있습니다. 기계학습 전문가가 데이터만으로 학습하는 데 사용하는 가장 강력한 도구 중 두 가지는 클러스터링과 차원 감소입니다.

- 클러스터링은 유사한 항목의 개별 버킷으로 데이터를 분할하는 프로세스입니다. 어떤 점에서 클러스터링은 해당 레이블을 몰라도 데이터를 분류하는 것과 같습니다. 예를 들어, 서가를 재구성 할 때 비슷한 장르를 함께 배치하거나 작성자의 성으로 그룹화 할 수 있습니다. 가장 널리 사용되는 클러스터링 알고리즘 중 하나는 K-means입니다. 

- 차원 성 감소는 훨씬 간단한 관점에서 데이터를 보기 위해 데이터를 조작하는 것입니다. 예를 들어 중복 된 특징을 없애면 저 차원 공간에서 동일한 데이터를 설명하고 어떤 기능이 중요한지 확인할 수 있습니다. 이 단순화는 데이터 시각화 또는 성능 효율성을위한 전처리에도 도움이됩니다. 최초의 알고리즘 중 하나는 PCA (Principle Component Analysis)이며, 일부는 새로운 기능인 자동 인코딩을 포함합니다 (7 장에서 다룰 것입니다).


### 1.4.3   Reinforcement Learning
- 감독과 감독되지 않은 학습은 교사의 존재가 모든 것임을 시사한다. 그러나 환경이 교사 역할을 하는 명확한 답변이 아니라 힌트를 제공하는 잘 학습 된 기계 학습의 영역이 있습니다. 학습 시스템은 올바른 방향으로 진행되고 있다는 구체적인 약속이 없는 상태에서 행동에 대한 피드백을 받습니다.

- 훈련 자료가 "선생님"에 의해 편리하게 분류되는 감독 학습과는 달리, 강화 학습은 환경이 조치에 어떻게 반응 하는지를 관찰함으로써 수집 된 정보를 교육합니다. 즉, 보강 학습은 환경과 상호 작용하여 어떤 조합의 행동이 가장 유리한 결과를 낳는지를 학습하는 일종의 기계 학습입니다. 우리는 이미 "환경"과 "행동"이라는 단어를 사용하여 알고리즘을 의인화하고 있기 때문에 일반적으로 시스템을 자율적인 "에이전트"라고 부릅니다. 따라서 이러한 유형의 기계 학습은 자연스럽게 로봇 영역으로 나타납니다.

- 환경의 에이전트에 대해 추론하기 위해 두 가지 새로운 개념, 즉 상태와 동작을 소개합니다. 특정 시간에 고정 된 세계의 상태를 state라고 합니다. 에이전트는 많은 동작 중 하나를 수행하여 현재 상태를 변경할 수 있습니다. 행동을 수행하도록 에이전트를 유도하기 위해 각 상태는 상응하는 보상을 산출합니다. 에이전트는 결국 state의 가치라고 불리는 각 국가의 예상되는 총 보상을 발견합니다.

- 다른 기계 학습 시스템과 마찬가지로 더 많은 데이터로 성능이 향상됩니다. 이 경우 데이터는 이전 경험의 기록입니다. 보강 학습에서 우리는 실행될 때까지 일련의 행동에 대한 최종 비용이나 보상을 알지 못합니다. 이러한 상황은 전통적인 감독 학습을 비효율적으로 만듭니다. 에이전트가 확실하게 알 수있는 유일한 정보는 이미 완료 한 일련의 작업 비용입니다. 이는 불완전합니다. 에이전트의 목표는 보상을 최대화하는 일련의 조치를 찾는 것입니다.


### 1.5   Existing Tools
- 기계학습 및 데이터 분석을 시작하는 가장 쉬운 방법 중 하나는 Scikit-learn Python 라이브러리 (http://scikit-learn.org/)를 사용하는 것입니다. 파이썬은 결국 산업 표준 구현이되는 아이디어를 프로토 타이핑하는 훌륭한 언어입니다. NumPy, SciPy 및 matplotlib과 같은 가장 영속적이고 성공적인 파이썬 라이브러리 중 일부는 Scikit 학습의 중추를 형성합니다. 도구는 간단하여 사용하기가 쉽습니다.

- 그러나 Scikit-learn은 라이브러리가 비교적 저수준이기 때문에 어셈블리 언어처럼 느낍니다. 결과적으로 정교한 알고리즘을 수행하면 쉽게 버그 코드가 발생할 수 있습니다. Scikit-learn과 직접 대화하는 대신 TensorFlow, Theano 또는 Caffe와 같은 고급 라이브러리를 사용하면 유연성을 희생하면서보다 강력한 설정을 제공합니다.

- Hadoop 또는 Apache Spark는 병렬 처리와 분산 컴퓨팅을 강조하는 큰 데이터를 처리하는 일부 고급 엔터프라이즈 표준 프레임 워크입니다. 일반적으로 이러한 병렬 아키텍처와 상호 작용하는 하위 레벨 라이브러리는 완전한 Java 인터페이스를 제공하는 Apache Mahout입니다.

- TensorFlow 외에도 가장 일반적인 기계 학습 라이브러리에는 그림 1.12와 같이 Theano, Caffe, Torch 및 Computational Graph Toolkit이 있습니다.

### 1.5.1 Theano
- 수년 동안 기계학습 종사자는 이미 Theano에서 잘 알려진 많은 신경망을 구현했습니다. 심볼릭 그래프를 사용하여 구현을 설계와 분리합니다. 
- 프로그래밍 환경은 Python으로 되어있어 초보자가 플랫폼에 관계없이 쉽게 시작할 수 있습니다.
- 낮은 수준의 인터페이스가 실질적인 사용자 정의를하기위한 지원은 거의 없습니다. 
- 또한 매번 이진 코드를 컴파일하여 신속하게 수정하거나 부담을 디버깅하기 때문에 신속한 프로토 타이핑에 상당한 오버 헤드가 있습니다.

### 1.5.2 Caffe
- Caffe의 기본 상호 작용은 단순한 Python 인터페이스로 인해 사용하기 쉽습니다. 
- 보다 복잡한 알고리즘이나 사용자 정의 된 신경 네트워크의 경우 C ++ 인터페이스를 사용할 수도 있습니다. 
- 대부분의 플랫폼은 C ++을 지원하기 때문에 쉽게 배포 할 수 있습니다. 
- 그러나 Caffe의 목적은 주로 이미지를 다루는 네트워크에 대한 것입니다. 텍스트 또는 시계열 데이터는 Caffe를 처리하는 데 자연스럽지 않습니다.

### 1.5.3 Torch
- 루아는 이 프레임 워크에서 선택한 프로그래밍 언어입니다. 
- Lua 환경은 대부분의 개발자들에게는 외국 적이기 때문에, 기계 학습 프로젝트를 위해 Torch를 사용하는 것과 관련된 사소한 위험이 있습니다. 
- 그러나 밝은 면에서 토치는 최적화 솔버를 강력하게 지원하므로 휠을 재발견 할 필요가 없습니다.

### 1.5.4 Computational Graph Toolkit
- 캘리포니아 대학 버클리 연구소 (University of California, Berkeley)는 기계 학습에 자주 사용되는 일반 그래프 작업을 위해 Computational Graph Toolkit (CGT)을 출시했습니다. Theano와 동일한 기능을 일부 지원하지만 병렬 처리에 중점을 둡니다. 
- 불행히도 문서는 다른 도서관에 비해 상대적으로 희박하며 커뮤니티는 Theano 또는 Caffe만큼 널리 퍼지지 않습니다.

### 1.6 TensorFlow
- Google은 2015 년 말에 Apache 2.0 라이선스에 따라 TensorFlow라는 기계 학습 프레임 워크를 공개했습니다. 그 전에는 Google이 음성 인식, 검색, 사진 및 Gmail과 같은 다른 응용 프로그램에서 독점적으로 사용했습니다.

- 이 라이브러리는 C ++로 구현되었으며 편리한 Python API와 낮은 C ++ API를 제공합니다. 종속성이 더 간단하기 때문에 TensorFlow를 다양한 아키텍처에 신속하게 배포 할 수 있습니다.

- Theano와 마찬가지로 계산은 설계를 구현과 분리하는 순서도로 설명됩니다. 별다른 번거 로움없이이 이분법은 수천 개의 GPU가있는 대규모 교육 시스템뿐만 아니라 모바일 장치에서도 동일한 디자인을 구현할 수 있습니다. 단일 시스템은 광범위한 플랫폼에 걸쳐 있습니다.

- TensorFlow의 가장 중요한 특성 중 하나는 자동 차별화 기능입니다. 많은 핵심 계산을 재정의 할 필요없이 새로운 네트워크를 실험 할 수 있습니다.

- 이 라이브러리의 또 다른 특징은 TensorBoard라는 대화 형 시각화 환경입니다. 이 도구는 데이터가 변환되는 방식에 대한 순서도를 보여주고 시간 경과에 따른 요약 로그를 표시하고 성능을 추적합니다. 그림 1.13은 TensorBoard가 사용중인 모습을 보여줍니다. 다음 장에서는 더 자세한 내용을 다룰 것입니다.

- Theano와는 달리 TensorFlow의 프로토 타이핑은 많은 작업이 미리 컴파일되기 때문에 훨씬 빠릅니다 (코드는 분 단위가 아닌 초 만에 시작됩니다). 서브 그래프 실행으로 인해 코드를 쉽게 디버깅 할 수 있습니다. 이것이 의미하는 바는 계산의 전체 세그먼트가 재 계산없이 재사용 될 수 있다는 것입니다.

- TensorFlow는 신경망뿐만 아니라 기본 매트릭스 계산 및 조작 도구도 갖추고 있습니다. Torch 나 Caffe와 같은 대부분의 라이브러리는 심 신경 네트워크 전용 으로 설계되었지만 TensorFlow는 확장 성과 확장 성이 뛰어납니다.

### 1.8   Summary
- TensorFlow는 기계 학습 솔루션을 구현하기 위해 전문가 및 연구원들 사이에서 선택의 도구가 되었습니다. 다음 장에서는 TensorFlow 인터페이스의 내부 동작에 대해 자세하게 설명합니다. 3 장에서 6 장은 TensorFlow가 고전 알고리즘을 구현하는 방법을 탐구하는 반면, 7 장에서 10 장은 현대 신경 네트워크 아키텍처에서 TensorFlow의 탁월함을 보여줍니다.

- 이 장에서는 다음을 포함하여 기계 학습에 대해 많은 것을 배웠습니다.

- 기계 학습은 새로운 입력에 대해 유용한 진술을 할 수있는 전문가 시스템을 개발하기 위해 예제를 사용하는 것에 관한 것입니다.
- ML의 핵심 속성은 더 많은 교육 데이터로 성능이 향상되는 경향이 있다는 것입니다.
- 수년 동안 학자들은 감독 학습, 감독되지 않은 학습 및 강화 학습과 같은 대부분의 문제에 부합되는 세 가지 주요 원형을 만들었습니다.
- 실제 세계 문제가 기계 학습 관점에서 공식화 된 후에 여러 알고리즘을 사용할 수있게됩니다. 구현을 수행하기위한 많은 소프트웨어 라이브러리와 프레임 워크 중에서 TensorFlow를 우리의 총알으로 선택했습니다. Google이 개발하고 번성 한 커뮤니티의 지원을받는 TensorFlow는 업계 표준 코드를 쉽게 구현할 수있는 방법을 제공합니다.
