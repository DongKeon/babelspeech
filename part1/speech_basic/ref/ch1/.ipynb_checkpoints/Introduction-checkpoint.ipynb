### 1.1 Automatic Speech Recognition: A Bridge for Better Communication
- 자동 음성 인식 (ASR)은 지난 50 년 동안 활발한 연구 분야였습니다. 
- 그것은 인간과 사람과 기계 간의 더 나은 커뮤니케이션을 육성하는 데 항상 중요한 다리로 간주되어 왔습니다.
- 그러나 과거에는 연설이 실제로 인간 - 기계 통신에서 중요한 양식이 아니었습니다. 
- 그 당시의 기술이 대부분의 실제 사용 조건 하에서 대부분의 실제 사용자에게 사용 가능한 막대를 전달하기에 충분하지 않았고, 많은 경우 키보드와 마우스 같은 대체 통신 방식이 통신 효율성에서 음성을 크게 능가했기 때문입니다.
- 최근 몇 년 동안 음성 기술은 우리가 살고 일하는 방식을 바꾸기 시작했으며 인간이 일부 장치와 상호 작용하는 주요 수단 중 하나가 되었습니다. 이러한 추세는 몇 가지 핵심 분야에서의 진보로 인해 시작되었습니다. 


    - 첫째, 무어의 법칙은 계속 기능합니다. 멀티 코어 프로세서, GPGPU (General Purpose Graphics Processing Unit) 및 CPU / GPU 클러스터를 통해 오늘날 사용할 수있는 컴퓨팅 성능은 불과 10 년 전의 것보다 몇 배 더 높습니다. 이것은 보다 강력하면서도 복잡한 모델의 학습을 가능하게 합니다. 계산이 필요한 모델은 ASR 시스템의 오류율을 크게 줄였습니다

    - 둘째, 인터넷과 클라우드 컴퓨팅의 지속적인 발전 덕분에 이전보다 더 많은 데이터에 액세스 할 수 있게 되었습니다. 실제 사용 시나리오에서 수집 한 큰 데이터에 모델을 구축함으로써 이전에 만들어진 많은 모델 가정을 제거하고 시스템을 보다 강력하게 만들 수 있습니다. 

    - 셋째, 모바일 기기, 웨어러블 기기, 지능형 거실 기기 및 차량 내 인포테인먼트 시스템이 보급되었습니다. 이러한 장치 및 시스템에서 키보드 및 마우스와 같은 대체 상호 작용 방식은 개인용 컴퓨터보다 쉽지 않습니다. 인간과 인간의 의사 소통의 자연스러운 방법 인 스피치와 대다수의 사람들이 이미 가지고 있는 기술은 보다 유리한 상호 작용이됩니다 이러한 장치 및 시스템의 양식. 최근에는 음성 기술이 중요한 역할을하는 많은 응용 프로그램이 있습니다. 

- 이러한 응용 프로그램은 HHC (Human-Human Communication)를 향상시키고 HMC (Human-Machine Communication)를 향상시키는 데 도움이 되는 응용 프로그램으로 분류 할 수 있습니다.


### 1.1.1 Human–Human Communication
- 음성 기술은 인간과 인간의 상호 작용을 가로막는 장벽을 제거 할 수 있습니다. 
- 과거에는 서로 다른 언어를 사용하는 사람들이 서로 대화 할 수있는 사람의 통역사가 필요했습니다. 
- 이것은 사람들이 누구와 통신 할 수 있고 언제 통신이 일어날 수 있는지에 대한 중요한 제한을 설정합니다. 예를 들어, 중국어를 말할 수없는 사람들은 종종 혼자서 중국을 여행하기가 어렵다고 생각합니다. 


### 1.1.2 Human–Machine Communication
- 음성 기술은 HMC를 크게 향상시킬 수 있습니다. 이 범주에서 가장 많이 사용되는 응용 프로그램에는 음성 검색, 개인용 정보 단말기, 게임, 거실 상호 작용 시스템 및 차량 내 인포테인먼트 시스템이 있습니다.
- 음성 검색 (VS) [24-26] 응용 프로그램을 사용하면 사용자가 식당, 운전 경로 및 제품 리뷰와 같은 정보를 음성으로 직접 검색 할 수 있습니다. 사용자가 검색 쿼리를 입력하는 데 필요한 노력이 크게 줄어 듭니다. 
- 요즘 음성 검색 응용 프로그램은 iPhone, Windows Phone 및 Android Phone과 같은 모바일 장치에서 널리 사용됩니다. PDA (Personal Digital Assistance)는 10 년 동안 프로토 타입으로 제작되었습니다. 하나, 애플이 아이폰에서 시리 (Siri) 시스템을 발표 한 후에야 인기를 얻었다. 그 이후로 많은 다른 회사들이 유사한 제품을 출시했습니다. 

- PDA는 귀하의 모바일 장치에 저장된 정보, 일부 세계 지식, 시스템과의 상호 작용 내역을 파악하여 사용자에게 더 나은 서비스를 제공 할 수 있습니다. 
- PDA가 할 수 있는 예제 작업에는 음성 명령을 직접 실행하여 전화 번호로 전화를 걸거나, 미팅을 예약하거나, 답변을 요청하거나, 음악을 검색하는 등의 작업이 포함됩니다.
- 사용자는 음악을 듣거나 정보를 요청하거나 시스템을 제어 할 수 있습니다. 그러나 이러한 시스템은 서로 다른 조건에서 사용되기 때문에 서로 다른 설계상의 어려움이 있습니다.
- 이 하위 절에서 논의 된 모든 응용 프로그램과 시스템은 음성 언어 시스템의 예입니다 [11]. 도 1.2에 도시 된 바와 같이, 음성 언어 시스템은 종종 4 가지 주요 구성 요소 중 하나 이상을 포함한다 : 음성을 텍스트로 변환하는 음성 
- 인식 구성 요소; 말로 된 단어에서 의미 론적 정보를 찾는 구어체 이해 구성 요소; 음성 정보를 전달하는 텍스트 음성 변환 (text-to-speech) 컴포넌트; 응용 프로그램 및 기타 세 가지 구성 요소와 통신하는 대화 상자 관리자입니다. 
- 이러한 모든 구성 요소는 성공적인 구어 시스템을 구축하는 데 중요합니다. 이 책에서는 ASR 구성 요소에만 초점을 맞춥니다.

### 1.2 ASR 시스템의 기본 아키텍처
- ASR 시스템의 일반적인 아키텍처는 그림 1.3에 나와 있습니다. 
- ASR 시스템은 신호 처리 및 특징 추출, 음향 모델 (AM), 언어 모델 (LM) 및 가설 검색의 네 가지 주요 구성 요소를 가지고 있습니다.
- 신호 처리 및 특징 추출 컴포넌트는 입력으로서 오디오 신호를 취하고, 잡음 및 채널 왜곡을 제거함으로써 음성을 강화하고, 신호를 시간 영역에서 주파수 영역으로 변환하고, 후속하는 음향 모델에 적합한 현저한 특징 벡터를 추출한다.
- 어쿠스틱 모델은 음향학 및 음성학에 대한 지식을 통합하고, 특징 추출 컴포넌트에서 생성 된 특징을 입력으로 받아, 가변 길이 특징 시퀀스에 대한 AM 스코어를 생성합니다. 
- 언어 모델은 (일반적으로 텍스트) 훈련 자료에서 단어 사이의 상관 관계를 학습하여 가정 된 단어 순서 또는 LM 점수의 확률을 추정합니다. 도메인 또는 작업에 대한 사전 지식이있는 경우 LM 점수를 더 정확하게 추정 할 수 있습니다. 
- 가설 탐색 컴포넌트는 특징 벡터 시퀀스 및 가정 된 단어 열을 추출하고, 가장 높은 점수를 갖는 단어 열을 인식 결과로서 출력한다. 이 책에서는 AM 구성 요소에 중점을 둡니다.
- AM 구성 요소가 처리해야 할 두 가지 주요 쟁점은 가변 길이 특징 벡터 및 오디오 신호의 가변성 가변 길이 특징 문제는 종종 동적 시간 왜곡 (DTW)과 은닉 마르코프 모델 (HMM) [18]과 같은 기술로 다루어집니다.

![img_1](./img/ch1_3.png)

- 오디오 신호의 가변성은 스피커 특성 (예 : 성별, 질병 또는 스트레스), 발화 양식 및 속도, 환경 소음, 사이드 토크, 채널 왜곡 (예 : 마이크 차이), 방언 차이, 그리고 비회원 악센트. 성공적인 음성 인식 시스템은 이러한 모든 음향 적 가변성과 대립해야합니다.

- 과거에는 지배적인 최첨단 ASR 시스템이 일반적으로 특징 벡터 및 가우시안으로 melfrequency cepstral coefficient(MFCC) [4] 또는 상대 스펙트럼 변환- 지각 선형 예측 (RASTA-PLP) [7] 혼합 모델 (GMM) -HMM을 음향 모델로 사용한다.
- 1990 년대에 이 GMM-HMM AM은 ML (maximum likelihood) 훈련 기준을 사용하여 훈련되었습니다. 
- 2000 년에는 MCE (Minimum Classification Error) [13]와 MPE (Minimum Phone Error) [16]와 같은 시퀀스 판별 훈련 알고리즘이 제안되었고 ASR 정확도가 더욱 향상되었다.
- 최근 수년간, 계산 능력의 지속적인 향상, 대규모 훈련 세트의 가용성, 그리고 이들 모델에 대한 더 나은 이해 덕분에 심 신경 네트워크 (DNN) [3, 9]와 같은 차별적 인 계층 적 모델이 실행 가능하고 오류율을 현저히 감소시켰다 . 예를 들어, 문맥 의존적 (CD) DNN-HMM은 시퀀스 판별 기준으로 훈련 된 종래의 GMM-HMMs에 비해 교환기 대화식 녹음 작업에서 오류율을 1/3로 줄였다.
- 이 책에서는 DNN, CNN (convolutional neural network) 및 RNN (recurrent neural network)을 포함하여 이러한 차별적 인 계층 모델의 신기술을 설명합니다. 우리는이 모델의 이론적 토대와 시스템을 작동시키는 실용적인 트릭을 논의합니다.

### 1.3 Book Organization
- 우리는 이 책을 다섯 부분으로 구성합니다. 기존의 GMM-HMM 시스템 뿐만 아니라 관련 수학 모델 및 변형에 대해 Part I을 정립하고 [5, 10, 19]의 확립 된 책과 [12, 17]의 튜토리얼 자료에서 내용을 추출합니다.
- 2 부에서는 DNN, 초기화 및 교육 알고리즘을 소개합니다. 
- 3 부에서는 음성 인식을 위한 DNN-HMM 하이브리드 시스템, DNN의 교육 및 평가 속도 향상을 위한 트릭, 시퀀스 차별적 교육 알고리즘에 대해 설명합니다.
- 파트 IV는 공동 표현 학습 및 모델 최적화 관점에서 DNN을 설명합니다. 
- 이 견해를 토대로 음성 인식을 위한 직렬 식 시스템과 DNN을 위한 적응 기법을 소개한다. 
- 파트 5에서는 다국어 DNN, 반복적 인 신경 네트워크 및 전산 네트워크와 같은 고급 모델을 제공합니다. 

### 1.3.1 제 1 부 : 기존의 어쿠스틱 모델
- 2 장과 3 장은 일반적인 GMM-HMM 음향 모델에 대한 기본적인 이론적 토대를 제공합니다. 이 장에서는 나머지 책에서 소개 할 차별적 인 계층 적 모델을 이해할 수 있도록 준비합니다.2 장에서는 가우시안 혼합 모델에 대해 논의한다. 확률 및 기대 최대화 [15] 학습 알고리즘을 사용합니다.
- 3 장 현대에 사용되는 가장 눈에 띄는 기술인 숨겨진 마르코프 모델을 소개합니다. 음성 인식 시스템. HMM이 가변 길이 신호 문제를 어떻게 처리 하는지를 설명하고,  역방향 학습 알고리즘과 비터 비 디코딩 알고리즘을 설명합니다 .GMM-HMM은 우리가 집중적으로 다루는  CD-DNN-HMM 시스템이 우세하기 전에 현대 음성 인식 시스템의 기초를 형성합니다 이 책.

### 1.3.2 파트 II : 딥 뉴럴 네트워크
- 4 장과 5 장에서는 심 신경 네트워크에 대해 심도있게 살펴 봅니다. 실제 시스템을 구축 할 때 잘 작동하는 것으로 입증 된 기술을 강조하고 이러한 기술이 이론적 관점과 실용적인 관점에서 어떻게 그리고 왜 작동하는지 자세히 설명합니다.

### 4 장에서는 심층 신경망, 유명한 역 전파
- 알고리즘 [14, 20], DNN을 효과적으로 훈련시키는 다양한 실습 효율적으로 5 장에서는 생성적인 사전 학습 (pretraining)과 차별적 사전 훈련 (discriminative pretraining) [21]을  포함한 고급 DNN 초기화 기술에 대해 논의합니다. 우리는 제한된 볼츠만 머신 (Boltzmann machine, RBM) [8],  시끄러운 자동 엔코더 [1], 그리고 둘 사이의 관계에 중점을 둔다.

### 1.3.3 파트 III : ASR 용 DNN-HMM 하이브리드 시스템
- 챕스에서. 6-8, 우리는 음성 인식을 위해 DNM과 HMM을 통합 할 때 검증 된 기법을 논의하기 위해 동일한 정신을 따릅니다. 6 장에서는 음성 신호의 연속 특성을 모델링하기 위해 HM을 사용하고 HMM에서 방출 확률을 모델링하기 위해 DNN을 사용하는 DNN-HMM 하이브리드 모델 [3]에 대해 설명한다. 7 장에서는 DNN-HMM 시스템의 교육 및 디코딩 속도를 향상시키는 실용적인 구현 방법에 대해 설명합니다. 8 장에서는 DNN-HMM 하이브리드 시스템의 인식 정확도를 더욱 향상시키는 시퀀스 차별적 훈련 알고리즘에 대해 설명합니다.

### 1.3.4 IV 부 : 딥 신경 회로망에서의 표상 학습
- 챕스에서. 9-11 우리는 DNN을 다른 각도에서 바라보고이 새로운 각도가 DNN이 음성 인식에 사용될 수있는 다른 접근 방법을 제안하는 방법에 대해 논의합니다. 9 장에서는 DNN의 공동 피쳐 학습 및 모델 최적화 관점을 설명합니다. 우리는 DNN이 임의의 
- 숨겨진 계층에서 분리 될 수 있으며 그 아래의 모든 계층을 특성 변환으로 간주하고 그 위에있는 모든 계층을 분류 모델로 간주합니다. 
- 10 장에서는 기존의 GMM-HMM에 포함 된 DNM을 별도의 피쳐 추출기로 사용하는 직렬 아키텍처 및 병목 시스템을 소개합니다. 
- 11 장에서는 DNN이 효과적으로 적용될 수있는 다양한 기술을 보여줍니다.

### 1.3.5 파트 V : 고급 심층 모델
- 챕스에서. 12에서 14까지 우리는 몇 가지 고급 심층 모델을 소개합니다. Chap. 12 우리는 DNN 기반 멀티 태스킹과 전송 학습을 설명합니다.
- 표현은 공유되고 관련 타스크로 전송됩니다. 우리는 이러한 기술을 시연하기 위해 공유 숨겨진 레이어 DNN 아키텍처를 사용하는 다국어 및 교차 언어 음성 인식을 주요 예로 사용합니다. Chap. 도 13은 음성 인식을위한 장시간 단기 기억 신경망을 포함하는 재발 성 신경망을 도시한다. 
- Chap. 14 우리는 전산 네트워크, 심근 신경망 (DNNs), 전산 신경망 (CNNs), 긴 단기 메모리 (LSTM)를 가진 버전을 포함하는 재발 성 신경망 (RNNs)과 같은 임의의 학습 기계를 설명하기위한 통일된 프레임 워크를 소개한다. 로지스틱 회귀 (logistic regression), 최대 엔트로피 모델 (maximum entropy model) 등이있다.
